```python
def coarse(file_list):
    # Read files from the file list and assign to a dictionary
    da = {}
    for i, file in enumerate(file_list):
        var_name = f"da{i}" 
        var_value = xr.open_dataset(file)
        da[var_name] = var_value
    
    # Coarsen the data to 1deg X 1 deg grid
    minx = -89.5
    maxx = 89.5
    miny = -179.5
    maxy = 179.5

    new_grid_x = np.arange(minx, maxx+1, 1)
    new_grid_y = np.arange(miny, maxy+1, 1)

    # interpolate using nearest neighbor (can use linear, etc. if desired)
    for d in da:
        # print(d)
        da[d] = da[d].interp(lon=new_grid_y, lat=new_grid_x, method="linear")  
        # print(da[d])
        da[d] = da[d][['CHL','DIATO','DINO','MICRO','NANO','PICO']]
    return da
```
